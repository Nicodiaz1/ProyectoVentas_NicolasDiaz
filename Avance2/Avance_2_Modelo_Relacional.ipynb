```python
import pandas as pd
import numpy as np
```

```python
df_ventas = pd.read_csv(r'C:\Users\kikid\Desktop\ProyectoVentas_NicolasDiaz\Avances\Avance1\ventasTransformed.csv')
```

```python
# Normalizar columna ciudadsucursal - llenar valores vacíos desde sucursal nombre
df_ventas['CiudadSucursal'] = df_ventas['CiudadSucursal'].fillna(
    df_ventas['SucursalNombre'].str.split(' ', n=1, expand=True)[1].str.split(' #', expand=True)[0]
)
# Limpiar CiudadSucursal: eliminar # y todo lo que viene después
df_ventas['CiudadSucursal'] = df_ventas['CiudadSucursal'].str.replace(r'\s*#.*', '', regex=True).str.strip()
# Normalizar SucursalNombre: normalizar el espacio antes de # (Techcore Bogota #2 -> Techcore Bogota#2)
df_ventas['SucursalNombre'] = df_ventas['SucursalNombre'].str.replace(' #', '#', regex=False)
```

```python
# Combinar columnas de año, mes y día en una columna de fecha
meses_dict = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}
df_ventas['Mes'] = df_ventas['Mes'].str.lower().map(meses_dict)
df_ventas['Día'] = df_ventas['Día'].astype(str).str.zfill(2)
df_ventas['Fecha'] = pd.to_datetime(
    df_ventas['Año'].astype(str) + '-' + df_ventas['Mes'].astype(str) + '-' + df_ventas['Día'].astype(str),
    format='%Y-%m-%d'
)
```

```python
df_ventas.head(10)
```

```python
# Extraer solo la hora de HoraVenta (eliminar la fecha 1899-12-30)
df_ventas['HoraVenta'] = pd.to_datetime(df_ventas['HoraVenta']).dt.strftime('%H:%M:%S')
```

```python
# Verificar duplicados
print(f"Filas totales: {len(df_ventas)}")
duplicados = df_ventas.duplicated().sum()
print(f"Número de filas duplicadas: {duplicados}")

```

```python
##CREACION DE DATA FRAMES PARA CADA TABLA

# 1. TABLA VENDEDORES
df_vendedores = df_ventas[['VendedorNombre']].drop_duplicates().reset_index(drop=True)
df_vendedores.insert(0, 'VendedorID', range(1, len(df_vendedores) + 1))
df_vendedores.columns = ['VendedorID', 'VendedorNombre']


print("1. TABLA VENDEDORES")
print(df_vendedores.head())
print(f"Total de registros: {len(df_vendedores)}\n")
```

```python
# 2. TABLA CLIENTES
df_clientes = df_ventas[['Cliente_Nombre', 'Suma de edadcliente', 'GeneroCliente', 'TelefonoCliente', 'EmailCliente', 'Direccion_Cliente']].drop_duplicates().reset_index(drop=True)
df_clientes.insert(0, 'ClienteID', range(1, len(df_clientes) + 1))
df_clientes.columns = ['ClienteID', 'Nombre', 'Edad', 'Género', 'Teléfono', 'Email', 'Dirección']


print("2. TABLA CLIENTES")
print(df_clientes.head())
print(f"Total de registros: {len(df_clientes)}\n")
```

```python
# 3. TABLA CIUDADES
df_ciudades = df_ventas[['CiudadSucursal']].drop_duplicates().reset_index(drop=True)
df_ciudades.insert(0, 'CiudadID', range(1, len(df_ciudades) + 1))
df_ciudades.columns = ['CiudadID', 'NombreCiudad']

print("3. TABLA CIUDADES")
print(df_ciudades.head())
print(f"Total de registros: {len(df_ciudades)}\n")
```

```python
# 4. TABLA SUCURSALES (con referencia a CiudadID)
df_sucursales = df_ventas[['SucursalNombre', 'CiudadSucursal']].drop_duplicates().reset_index(drop=True)
df_sucursales.insert(0, 'SucursalID', range(1, len(df_sucursales) + 1))
df_sucursales.columns = ['SucursalID', 'SucursalNombre', 'CiudadSucursal']

# Crear mapa de Ciudad a CiudadID
ciudad_map = dict(zip(df_ciudades['NombreCiudad'], df_ciudades['CiudadID']))
df_sucursales['CiudadID'] = df_sucursales['CiudadSucursal'].map(ciudad_map)

# Reorganizar columnas
df_sucursales = df_sucursales[['SucursalID', 'SucursalNombre', 'CiudadID']]


print("4. TABLA SUCURSALES")
print(df_sucursales.head(10))
print(f"Total de registros: {len(df_sucursales)}\n")
```

```python
# 5. TABLA PRODUCTOS (consolidar de las 3 columnas de productos)
productos_list = []

for i in range(1, 4):
    nombre_col = f'NombreProducto{i}'
    marca_col = f'MarcaProducto{i}'
    precio_col = f'Suma de PrecioUnitarioProducto{i}'
    
    # Filtrar solo filas donde existe el producto
    temp = df_ventas[[nombre_col, marca_col, precio_col]].dropna()
    
    if len(temp) > 0:
        temp_df = temp.copy()
        temp_df.columns = ['Nombre', 'Marca', 'Precio']
        productos_list.append(temp_df)

# Consolidar y eliminar duplicados
df_productos = pd.concat(productos_list, ignore_index=True).drop_duplicates().reset_index(drop=True)
df_productos.insert(0, 'ProductoID', range(1, len(df_productos) + 1))

# Reordenar columnas
df_productos = df_productos[['ProductoID', 'Nombre', 'Marca', 'Precio']]

print("5. TABLA PRODUCTOS")
print(df_productos.head(40))
print(f"Total de registros: {len(df_productos)}\n")
```

```python
# 6. TABLA FACTURAS (correspondiente a VENTA)
# Crear mapeos de nombres a IDs
vendedor_map = dict(zip(df_vendedores['VendedorNombre'], df_vendedores['VendedorID']))
sucursal_map = dict(zip(df_sucursales['SucursalNombre'], df_sucursales['SucursalID']))
cliente_map = dict(zip(df_clientes['Nombre'], df_clientes['ClienteID']))

df_facturas = df_ventas[[
    'Fecha', 
    'VendedorNombre', 'SucursalNombre', 'Cliente_Nombre', 'MetodoPago'
]].copy().reset_index(drop=True)

# Insertar FacturaID 
df_facturas.insert(0, 'FacturaID', range(1, len(df_facturas) + 1))

# Mapear nombres a IDs
df_facturas['VendedorID'] = df_facturas['VendedorNombre'].map(vendedor_map)
df_facturas['SucursalID'] = df_facturas['SucursalNombre'].map(sucursal_map)
df_facturas['ClienteID'] = df_facturas['Cliente_Nombre'].map(cliente_map)

# Reorganizar y seleccionar solo columnas requeridas
df_facturas = df_facturas[[
    'FacturaID', 'Fecha', 'SucursalID', 'ClienteID', 'VendedorID', 'MetodoPago'
]]

print("6. TABLA FACTURAS")
print(df_facturas.head(10000))
print(f"Total de registros: {len(df_facturas)}\n")
```

```python
# 7. TABLA DETALLE_FACTURAS (Tabla Asociativa - DETALLE_VENTA)
# Crear mapa de productos por nombre y marca
producto_map = {}
for _, row in df_productos.iterrows():
    key = (row['Nombre'], row['Marca'])
    producto_map[key] = row['ProductoID']

detalle_factura_list = []

for idx, row in df_ventas.iterrows():
    factura_id = idx + 1  # Corresponde con FacturaID
    descuento = row['Suma de DescuentoVenta']  # Descuento de la factura
    
    # Procesar hasta 3 productos por factura
    for i in range(1, 4):
        nombre_col = f'NombreProducto{i}'
        marca_col = f'MarcaProducto{i}'
        cantidad_col = f'Suma de CantidadProducto{i}'
        
        # Verificar si el producto existe en esta factura
        if pd.notna(row[nombre_col]):
            producto_key = (row[nombre_col], row[marca_col])
            if producto_key in producto_map:
                detalle_factura_list.append({
                    'FacturaID': factura_id,
                    'ProductoID': producto_map[producto_key],
                    'Cantidad': row[cantidad_col],
                    'Descuento': descuento
                })

df_detalle_facturas = pd.DataFrame(detalle_factura_list)

# Insertar DetalleID como clave primaria
df_detalle_facturas.insert(0, 'DetalleID', range(1, len(df_detalle_facturas) + 1))

print("7. TABLA DETALLE_FACTURAS")
print(df_detalle_facturas.head(10))
print(f"Total de registros: {len(df_detalle_facturas)}\n")
```

```python
# RESUMEN Y VALIDACIÓN DE INTEGRIDAD REFERENCIAL
print("RESUMEN - DATAFRAMES NORMALIZADOS CREADOS")

print(f'VENDEDORES: {len(df_vendedores)}')
print(f'CLIENTES: {len(df_clientes)}')
print(f'CIUDADES: {len(df_ciudades)}')
print(f'SUCURSALES: {len(df_sucursales)}')
print(f'PRODUCTOS: {len(df_productos)}')
print(f'FACTURAS: {len(df_facturas)}')
print(f'DETALLE_FACTURAS: {len(df_detalle_facturas)}\n')

print("\n")
print("VALIDACIÓN DE INTEGRIDAD REFERENCIAL")
print("\n")

# Validar FACTURAS -> VENDEDORES
vendedor_check = df_facturas['VendedorID'].isin(df_vendedores['VendedorID']).all()
print(f" FACTURAS.VendedorID → VENDEDORES.VendedorID: {vendedor_check}")

# Validar FACTURAS -> CLIENTES
cliente_check = df_facturas['ClienteID'].isin(df_clientes['ClienteID']).all()
print(f" FACTURAS.ClienteID → CLIENTES.ClienteID: {cliente_check}")

# Validar FACTURAS -> SUCURSALES
sucursal_check = df_facturas['SucursalID'].isin(df_sucursales['SucursalID']).all()
print(f" FACTURAS.SucursalID → SUCURSALES.SucursalID: {sucursal_check}")

# Validar SUCURSALES -> CIUDADES
ciudad_check = df_sucursales['CiudadID'].isin(df_ciudades['CiudadID']).all()
print(f" SUCURSALES.CiudadID → CIUDADES.CiudadID: {ciudad_check}")

# Validar DETALLE_FACTURAS -> FACTURAS
detalle_factura_check = df_detalle_facturas['FacturaID'].isin(df_facturas['FacturaID']).all()
print(f" DETALLE_FACTURAS.FacturaID → FACTURAS.FacturaID: {detalle_factura_check}")

# Validar DETALLE_FACTURAS -> PRODUCTOS
detalle_producto_check = df_detalle_facturas['ProductoID'].isin(df_productos['ProductoID']).all()
print(f" DETALLE_FACTURAS.ProductoID → PRODUCTOS.ProductoID: {detalle_producto_check}")
```

```python
print("ANÁLISIS DE DATOS")

# 1. TOTAL DE VENTAS POR MARCA
print("\nTOTAL DE VENTAS POR MARCA")

# Unir detalle_facturas con productos para obtener marca y cantidad
ventas_por_marca = df_detalle_facturas.merge(df_productos[['ProductoID', 'Marca']], on='ProductoID')
ventas_por_marca_total = ventas_por_marca.groupby('Marca')['Cantidad'].sum().sort_values(ascending=False)

print(ventas_por_marca_total)
print(f"\nTotal de unidades vendidas: {ventas_por_marca_total.sum():.0f}")

```

```python
# 2. TOP 10 PRODUCTOS MÁS VENDIDOS
print("TOP 10 PRODUCTOS MÁS VENDIDOS")

# Agrupar detalles por producto y sumar cantidad
top_productos = df_detalle_facturas.groupby('ProductoID')['Cantidad'].sum().sort_values(ascending=False).head(10)

# Unir con información de productos
top_productos_info = top_productos.reset_index()
top_productos_info = top_productos_info.merge(df_productos[['ProductoID', 'Nombre', 'Marca', 'Precio']], on='ProductoID')
top_productos_info.columns = ['ProductoID', 'Cantidad Vendida', 'Nombre', 'Marca', 'Precio Unitario']

# Reordenar columnas para mejor visualización
top_productos_info = top_productos_info[['Nombre', 'Marca', 'Cantidad Vendida', 'Precio Unitario', 'ProductoID']]


print(top_productos_info.to_string(index=False))
print(f"\nTotal de unidades en Top 10: {top_productos_info['Cantidad Vendida'].sum():.0f}")
```

```python
#PRODUCTOS SIN VENTAS
print("PRODUCTOS SIN VENTAS")

# Obtener productos que se vendieron
productos_vendidos = set(df_detalle_facturas['ProductoID'].unique())

# Obtener todos los productos
todos_productos = set(df_productos['ProductoID'].unique())

# Productos sin ventas
productos_sin_ventas = todos_productos - productos_vendidos

if len(productos_sin_ventas) > 0:
    print(f"productos sin ventas: {len(productos_sin_ventas)} \n")
    
    productos_sin_ventas_info = df_productos[df_productos['ProductoID'].isin(productos_sin_ventas)].sort_values('ProductoID')
    print(productos_sin_ventas_info.to_string(index=False))
else:
    print(f"Todos los {len(todos_productos)} productos tienen ventas registradas")

print(f"\nProductos vendidos: {len(productos_vendidos)}")
print(f"Productos totales: {len(todos_productos)}")
print(f"Tasa de venta: {(len(productos_vendidos)/len(todos_productos)*100):.1f}%")
```
